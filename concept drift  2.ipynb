```python
def add_indicators(df):
    
    df["SMA_10"] = df["Close"].rolling(window=10).mean()
    df["SMA_20"] = df["Close"].rolling(window=20).mean()
    df["EMA_10"] = df["Close"].ewm(span=10, adjust=False).mean()
    
    
    delta = df["Close"].diff()
    gain = np.where(delta > 0, delta, 0)
    loss = np.where(delta < 0, -delta, 0)
    avg_gain = pd.Series(gain).rolling(window=14).mean()
    avg_loss = pd.Series(loss).rolling(window=14).mean()
    rs = avg_gain / avg_loss
    df["RSI"] = 100 - (100 / (1 + rs))
    
    
    short_ema = df["Close"].ewm(span=12, adjust=False).mean()
    long_ema = df["Close"].ewm(span=26, adjust=False).mean()
    df["MACD"] = short_ema - long_ema
    df["Signal_Line"] = df["MACD"].ewm(span=9, adjust=False).mean()
    
    
    df["BB_Middle"] = df["Close"].rolling(window=20).mean()
    df["BB_Upper"] = df["BB_Middle"] + (df["Close"].rolling(window=20).std() * 2)
    df["BB_Lower"] = df["BB_Middle"] - (df["Close"].rolling(window=20).std() * 2)
    
    
    df["Momentum"] = df["Close"] - df["Close"].shift(10)
    
    
    df["Volatility"] = df["Close"].rolling(window=10).std()
    
    return df
```

```python
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
from torch.utils.data import DataLoader, TensorDataset
```

```python
import ta

for df in [df_2016, df_2021]:
    df['return'] = df['Close'].pct_change()
    df['rsi'] = ta.momentum.RSIIndicator(df['Close'], window=14).rsi()
    df['ema_10'] = ta.trend.EMAIndicator(df['Close'], window=10).ema_indicator()
    df['ema_30'] = ta.trend.EMAIndicator(df['Close'], window=30).ema_indicator()
    df['volatility'] = ta.volatility.BollingerBands(df['Close'], window=20).bollinger_hband_indicator()
    df.fillna(0, inplace=True)

features = ['Open','High','Low','Close','return','rsi','ema_10','ema_30','volatility']

# Create up/down label (1 if next close > current close, else 0)
df_2016['Updown'] = (df_2016['Close'].shift(-4) > df_2016['Close']).astype(int)
df_2021['Updown'] = (df_2021['Close'].shift(-4) > df_2021['Close']).astype(int)
df_2016.dropna(inplace=True)
df_2021.dropna(inplace=True)

# Scale the features
scaler = StandardScaler()
df_2016[features] = scaler.fit_transform(df_2016[features])
df_2021[features] = scaler.transform(df_2021[features])
```

```python
def create_sequences(data, seq_length=30):
    xs, ys = [], []
    for i in range(len(data) - seq_length):
        x = data[i:(i+seq_length), :-1]
        y = data[i+seq_length, -1]
        xs.append(x)
        ys.append(y)
    return np.array(xs), np.array(ys)

seq_len = 30
train_data = df_2016[features + ['Updown']].values
test_data = df_2021[features + ['Updown']].values

X_train, y_train = create_sequences(train_data, seq_len)
X_test, y_test = create_sequences(test_data, seq_len)

X_train_tensor = torch.tensor(X_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)
X_test_tensor = torch.tensor(X_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)

train_loader = DataLoader(TensorDataset(X_train_tensor, y_train_tensor), batch_size=64, shuffle=True)
test_loader = DataLoader(TensorDataset(X_test_tensor, y_test_tensor), batch_size=64, shuffle=False)
```

```python
# class LSTMClassifier(nn.Module):
#     def __init__(self, input_size, hidden_size, num_layers, output_size, dropout=0.2):
#         super(LSTMClassifier, self).__init__()
#         self.hidden_size = hidden_size
#         self.num_layers = num_layers

#         self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout)
#         self.fc = nn.Linear(hidden_size, output_size)
#         self.relu = nn.ReLU()
#         self.dropout = nn.Dropout(dropout)

#     def forward(self, x):
#         h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
#         c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

#         out, _ = self.lstm(x, (h0, c0))
#         out = self.dropout(out[:, -1, :])
#         out = self.fc(self.relu(out))
#         return out


class LSTMClassifier(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size, dropout=0.2):
        super(LSTMClassifier, self).__init__()
        self.hidden_size = hidden_size
        self.num_layers = num_layers

        
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True, dropout=dropout)
        self.bn = nn.BatchNorm1d(hidden_size)
        self.fc = nn.Linear(hidden_size, output_size)
        self.relu = nn.ReLU()
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        h0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)
        c0 = torch.zeros(self.num_layers, x.size(0), self.hidden_size).to(x.device)

        out, _ = self.lstm(x, (h0, c0))

        
        out = out[:, -1, :]              # shape: [batch_size, hidden_size]
        out = self.bn(out)               # normalize across batch
        out = self.relu(out)
        out = self.dropout(out)
        out = self.fc(out)
        return out

```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

input_size = len(features)
hidden_size = 128
num_layers = 2
output_size = 2
epochs = 200
learning_rate = 0.01
weight_decay = 1e-4

model = LSTMClassifier(input_size, hidden_size, num_layers, output_size).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate, weight_decay = weight_decay)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=epochs)
```

```python
train_losses = []

for epoch in range(epochs):
    model.train()
    total_loss = 0
    for X_batch, y_batch in train_loader:
        X_batch, y_batch = X_batch.to(device), y_batch.to(device)
        optimizer.zero_grad()
        outputs = model(X_batch)
        loss = criterion(outputs, y_batch)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()

    scheduler.step()
    
    avg_loss = total_loss / len(train_loader)
    train_losses.append(avg_loss)
    print(f"Epoch [{epoch+1}/{epochs}], Loss: {avg_loss:.4f}")

plt.plot(train_losses)
plt.title("Training Loss Curve")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.show()
```

```python
def evaluate_model(model, data_loader):
    model.eval()
    preds, true_labels = [], []
    with torch.no_grad():
        for X_batch, y_batch in data_loader:
            X_batch, y_batch = X_batch.to(device), y_batch.to(device)
            outputs = model(X_batch)
            _, predicted = torch.max(outputs, 1)
            preds.extend(predicted.cpu().numpy())
            true_labels.extend(y_batch.cpu().numpy())
    return np.array(true_labels), np.array(preds)


y_true_train, y_pred_train = evaluate_model(model, train_loader)
print("\nTraining (2016) Report:")
print(classification_report(y_true_train, y_pred_train, digits=4))


y_true_test, y_pred_test = evaluate_model(model, test_loader)
print("\nTesting (2021) Report:")
print(classification_report(y_true_test, y_pred_test, digits=4))
```

```python
train_acc = accuracy_score(y_true_train, y_pred_train)
test_acc = accuracy_score(y_true_test, y_pred_test)

plt.bar(['2016 (Train)', '2021 (Test)'], [train_acc, test_acc], color=['green', 'red'])
plt.title("Concept Drift: Accuracy Comparison")
plt.ylabel("Accuracy")
plt.show()

print(f"Training Accuracy (2016): {train_acc:.3f}")
print(f"Testing Accuracy (2021): {test_acc:.3f}")
```

```python
#model
import torch
import torch.nn as nn

class Adapter(nn.Module):
    def __init__(self, dim, bottleneck=32):
        super().__init__()
        self.down = nn.Linear(dim, bottleneck)
        self.up = nn.Linear(bottleneck, dim)
        self.act = nn.ReLU()
        self.scale = nn.Parameter(torch.tensor(0.1))

    def forward(self, x):
        # Low-rank adaptation path
        return self.scale * self.up(self.act(self.down(x)))

class TransformerClassifier(nn.Module):
    def __init__(self, input_size, num_heads=4, num_layers=2, hidden_dim=128, output_size=2, dropout=0.2, adapter_dim=32):
        super().__init__()
        self.embedding = nn.Linear(input_size, hidden_dim)

        # Transformer Encoder Layers
        self.encoder_layers = nn.ModuleList([
            nn.TransformerEncoderLayer(
                d_model=hidden_dim, nhead=num_heads, dim_feedforward=hidden_dim*2,
                dropout=dropout, batch_first=True
            ) for _ in range(num_layers)
        ])

        # Corresponding Adapters for each layer
        self.adapters = nn.ModuleList([
            Adapter(hidden_dim, adapter_dim) for _ in range(num_layers)
        ])

        # Final classification head
        self.fc = nn.Sequential(
            nn.LayerNorm(hidden_dim),
            nn.ReLU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, output_size)
        )

    def forward(self, x):
        # x shape: (batch, seq_len, input_size)
        x = self.embedding(x)
        for i, layer in enumerate(self.encoder_layers):
            residual = x
            x = layer(x)
            x = x + self.adapters[i](residual)  # Adapter residual connection
        x = x[:, -1, :]  # Use the last token
        return self.fc(x)
```

```python
# HELPERS
def create_sequences(df, features, seq_len=30):
    
    arr = df[features + ['Target']].values
    X, y = [], []
    for i in range(len(arr) - seq_len):
        X.append(arr[i:i+seq_len, :-1])
        y.append(arr[i+seq_len, -1])
    return np.array(X, dtype=np.float32), np.array(y, dtype=np.int64)


def build_loader(df, features, seq_len, batch_size, shuffle=False):
    
    X, y = create_sequences(df, features, seq_len)
    ds = TensorDataset(torch.from_numpy(X), torch.from_numpy(y))
    return DataLoader(ds, batch_size=batch_size, shuffle=shuffle, drop_last=True)


def train_model(model, loader, epochs, lr=1e-3, clip=1.0):
    
    crit = nn.CrossEntropyLoss()
    opt = torch.optim.AdamW(model.parameters(), lr=lr, weight_decay=1e-4)
    
    model.train()
    for ep in range(epochs):
        total_loss = 0.0
        for Xb, yb in loader:
            if torch.isnan(Xb).any():  # skip bad batches
                continue
            Xb, yb = Xb.to(device), yb.to(device)
            opt.zero_grad()
            out = model(Xb)
            loss = crit(out, yb)
            if torch.isnan(loss):
                continue
            loss.backward()
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)  # prevents exploding grads
            opt.step()
            total_loss += loss.item()

        if (ep + 1) % 10 == 0:
            print(f"Epoch {ep+1}/{epochs}, Loss: {total_loss/len(loader):.4f}")
    return model


def evaluate(model, loader):
   
    model.eval()
    preds, trues = [], []
    with torch.no_grad():
        for Xb, yb in loader:
            Xb = Xb.to(device)
            out = model(Xb)
            p = out.argmax(dim=1).cpu().numpy()
            preds.append(p)
            trues.append(yb.numpy())
    preds = np.concatenate(preds)
    trues = np.concatenate(trues)
    return accuracy_score(trues, preds)
```

```python
df = pd.read_csv(file_path, parse_dates=["Datetime"])
features = [c for c in df.columns if c not in ("Datetime", "Target")]
print(f"Using {len(features)} features: {features[:10]}...")

# Normalize using 2016 as baseline
scaler = StandardScaler()
scaler.fit(df[df['Datetime'].dt.year == 2016][features])
df[features] = scaler.transform(df[features])
```

```python
# ---------------- EXPERIMENT 1: STATIC TRANSFORMER ----------------
train2016 = df[df['Datetime'].dt.year == 2016]
test2021 = df[df['Datetime'].dt.year == 2021]
loader_train = build_loader(train2016, features, seq_len, batch_size, True)
loader_test = build_loader(test2021, features, seq_len, batch_size, False)

model_static = TransformerClassifier(len(features), num_heads, num_layers, hidden_dim, 2, dropout).to(device)
model_static = train_model(model_static, loader_train, epochs_static)
acc_train = evaluate(model_static, loader_train)
acc_test = evaluate(model_static, loader_test)
print(f"\nStatic Transformer Acc — Train(2016): {acc_train:.3f}, Test(2021): {acc_test:.3f}")
```

```python
# ---------------- EXPERIMENT 2: SLIDING WINDOW ----------------
years = [2016, 2017, 2018, 2019, 2020, 2021]

window = 2  # train on last 2 years
acc_sw = []

for y in years:
    window_years = list(range(y - window, y))
    train_window = df[df['Datetime'].dt.year.isin(window_years)]
    test_year = df[df['Datetime'].dt.year == y]
    if len(train_window) < 500 or len(test_year) < 500:
        continue

    model_sw = TransformerClassifier(
        len(features), num_heads, num_layers, hidden_dim, 2, dropout
    ).to(device)

    loader_train_w = build_loader(train_window, features, seq_len, batch_size, True)
    loader_test_y = build_loader(test_year, features, seq_len, batch_size, False)

    model_sw = train_model(model_sw, loader_train_w, epochs=epochs_update)
    acc_y = evaluate(model_sw, loader_test_y)
    acc_sw.append((y, acc_y))
    print(f"Year {y}: Sliding Window Accuracy = {acc_y:.3f}")
```

```python
# --- incremental yearly fine-tune ---
model_ft = TransformerClassifier(len(features), num_heads, num_layers,
                                 hidden_dim, 2, dropout).to(device)
model_ft = train_model(model_ft, loader_train, epochs_static, lr=1e-3)

accs_ft = []
for y in years:
    chunk = df[df['Datetime'].dt.year == y]
    loader_chunk = build_loader(chunk, features, seq_len, batch_size, True)
    acc_before = evaluate(model_ft, loader_chunk)

    # new optimizer each year + smaller lr
    model_ft = train_model(model_ft, loader_chunk, epochs=10, lr=1e-4)
    acc_after = evaluate(model_ft, loader_chunk)

    accs_ft.append((y, acc_before, acc_after))
    print(f"Year {y}: Before {acc_before:.3f}, After Fine-tune {acc_after:.3f}")
```

```python
# Updated metrics + safe-eval script
import numpy as np
import pandas as pd
import os
import json
import warnings
from sklearn.metrics import (
    accuracy_score,
    precision_recall_fscore_support,
    classification_report
)

# ---------------- CONFIG ----------------
force_load_npy = False   # set True only if you want to load pre-saved .npy files
npy_files = {
    "y_true_2016": "y_true_2016.npy",
    "y_pred_static_2016": "y_pred_static_2016.npy",
    "y_true_2021": "y_true_2021.npy",
    "y_pred_static_2021": "y_pred_static_2021.npy",
    "y_pred_sliding_2021": "y_pred_sliding_2021.npy",
    "y_pred_finetune_2021": "y_pred_finetune_2021.npy"
}
label_names = ["Down", "Up"]   # 0 -> Down, 1 -> Up
labels = [0, 1]

# ---------------- HELPERS ----------------
def compute_scores(y_true, y_pred, labels=[0,1], label_names=None, zero_division=0):
    acc = accuracy_score(y_true, y_pred)
    prec, rec, f1, support = precision_recall_fscore_support(
        y_true, y_pred, labels=labels, zero_division=zero_division
    )
    prf_macro = precision_recall_fscore_support(y_true, y_pred, average='macro', zero_division=zero_division)
    prf_micro = precision_recall_fscore_support(y_true, y_pred, average='micro', zero_division=zero_division)
    prf_weighted = precision_recall_fscore_support(y_true, y_pred, average='weighted', zero_division=zero_division)
    return {
        "accuracy": float(acc),
        "precision_per_class": prec.tolist() if hasattr(prec, "tolist") else list(prec),
        "recall_per_class": rec.tolist() if hasattr(rec, "tolist") else list(rec),
        "f1_per_class": f1.tolist() if hasattr(f1, "tolist") else list(f1),
        "support_per_class": support.tolist() if hasattr(support, "tolist") else list(support),
        "precision_macro": float(prf_macro[0]),
        "recall_macro": float(prf_macro[1]),
        "f1_macro": float(prf_macro[2]),
        "precision_micro": float(prf_micro[0]),
        "recall_micro": float(prf_micro[1]),
        "f1_micro": float(prf_micro[2]),
        "precision_weighted": float(prf_weighted[0]),
        "recall_weighted": float(prf_weighted[1]),
        "f1_weighted": float(prf_weighted[2]),
        "classification_report_text": classification_report(y_true, y_pred, labels=labels, target_names=label_names, digits=4, zero_division=0)
    }

def safe_eval_predictions(model_obj, loader_obj):
    """
    Evaluate model on loader and return (y_true, y_pred) numpy arrays.
    Works with either CPU or CUDA models.
    """
    import torch
    if model_obj is None or loader_obj is None:
        return np.array([]), np.array([])

    device = next(model_obj.parameters()).device if any(True for _ in model_obj.parameters()) else torch.device("cpu")
    model_obj.eval()
    y_trues = []
    y_preds = []
    with torch.no_grad():
        for Xb, yb in loader_obj:
            # move inputs to model device (if tensors)
            if isinstance(Xb, torch.Tensor):
                Xb = Xb.to(device)
            out = model_obj(Xb)
            preds = out.argmax(dim=1).cpu().numpy()
            # yb might be tensor or numpy
            if isinstance(yb, torch.Tensor):
                yb_arr = yb.cpu().numpy()
            else:
                yb_arr = np.asarray(yb)
            y_preds.append(preds)
            y_trues.append(yb_arr)
    if len(y_preds) == 0:
        return np.array([]), np.array([])
    y_true = np.concatenate(y_trues)
    y_pred = np.concatenate(y_preds)
    return y_true, y_pred

# ---------------- LOAD or BUILD arrays ----------------
arrays = {}

if force_load_npy:
    # load expected npy files
    for k, f in npy_files.items():
        if os.path.exists(f):
            arrays[k] = np.load(f, allow_pickle=True)
        else:
            arrays[k] = None
else:
    # look for in-memory variables created earlier in the notebook
    env = globals()
    candidate_map = {
        "y_true_train": "y_true_2016",
        "y_pred_train": "y_pred_static_2016",
        "y_true_test": "y_true_2021",
        "y_pred_test": "y_pred_static_2021",
        "y_pred_sliding_2021": "y_pred_sliding_2021",
        "y_pred_finetune_2021": "y_pred_finetune_2021",
        # exact names too
        "y_true_2016": "y_true_2016",
        "y_pred_static_2016": "y_pred_static_2016",
        "y_true_2021": "y_true_2021",
        "y_pred_static_2021": "y_pred_static_2021"
    }
    for varname, key in candidate_map.items():
        if varname in env:
            arrays[key] = env[varname]

    # If critical arrays missing, try to evaluate models/loaders present in memory
    need = [k for k in ["y_true_2016","y_pred_static_2016","y_true_2021","y_pred_static_2021"] if k not in arrays]
    # Try common variable names from your notebook
    # (rebuild loaders with drop_last=False if build_loader exists in env)
    if "build_loader" in env and "df" in env and "features" in env:
        # try to (re)create train/test loaders with drop_last=False by calling build_loader with updated behavior
        try:
            # some build_loader functions close over drop_last; attempt to create a fresh one here:
            from torch.utils.data import DataLoader, TensorDataset
            def build_loader_local(df_in, features_in, seq_len=30, batch_size=64, shuffle=False):
                arr = df_in[features_in + ["Target"]].values
                X, y = [], []
                for i in range(len(arr) - seq_len):
                    X.append(arr[i:i+seq_len, :-1].astype(np.float32))
                    y.append(arr[i+seq_len, -1].astype(np.int64))
                X = np.array(X)
                y = np.array(y)
                ds = TensorDataset(torch.from_numpy(X), torch.from_numpy(y))
                return DataLoader(ds, batch_size=batch_size, shuffle=shuffle, drop_last=False)
            # create loaders only if train/test split available by year
            if "df" in env and hasattr(env["df"], "Datetime"):
                df_local = env["df"]
                train2016_local = df_local[df_local["Datetime"].dt.year == 2016]
                test2021_local = df_local[df_local["Datetime"].dt.year == 2021]
                if len(train2016_local) > 300:
                    env["train_loader"] = build_loader_local(train2016_local, env["features"], seq_len=env.get("seq_len",30), batch_size=env.get("batch_size",64), shuffle=True)
                if len(test2021_local) > 300:
                    env["test_loader"] = build_loader_local(test2021_local, env["features"], seq_len=env.get("seq_len",30), batch_size=env.get("batch_size",64), shuffle=False)
        except Exception as e:
            print("Could not rebuild loaders locally:", e)

    # Evaluate models on available loaders
    eval_try = [
        ("model", "train_loader", "y_true_2016", "y_pred_static_2016"),
        ("model", "test_loader", "y_true_2021", "y_pred_static_2021"),
        ("model_static", "loader_train", "y_true_2016", "y_pred_static_2016"),
        ("model_static", "loader_test", "y_true_2021", "y_pred_static_2021"),
        ("model_static", "test_loader", "y_true_2021", "y_pred_static_2021"),
        ("model_sw", "loader_test_y", None, "y_pred_sliding_2021"),
        ("model_sw", "loader_test", None, "y_pred_sliding_2021"),
        ("model_ft", "loader_test_y", None, "y_pred_finetune_2021"),
        ("model_ft", "loader_test", None, "y_pred_finetune_2021"),
    ]
    for model_name, loader_name, key_true, key_pred in eval_try:
        if key_pred in arrays and arrays.get(key_pred) is not None:
            continue
        if model_name in env and loader_name in env:
            try:
                y_true_tmp, y_pred_tmp = safe_eval_predictions(env[model_name], env[loader_name])
                if y_true_tmp.size > 0:
                    if key_true is not None:
                        arrays[key_true] = y_true_tmp
                        arrays[key_pred] = y_pred_tmp
                    else:
                        # loader yields test y_true for 2021 usually; use it if present
                        if "2021" in loader_name or "test" in loader_name:
                            arrays["y_true_2021"] = y_true_tmp
                            arrays[key_pred] = y_pred_tmp
                        elif "train" in loader_name:
                            arrays["y_true_2016"] = y_true_tmp
                            arrays[key_pred] = y_pred_tmp
                    print(f"Evaluated {model_name} on {loader_name} -> saved to arrays.")
            except Exception as e:
                print(f"Failed to eval {model_name} on {loader_name}: {e}")

# Check required arrays presence
required = ["y_true_2016", "y_pred_static_2016", "y_true_2021", "y_pred_static_2021"]
missing = [k for k in required if k not in arrays or arrays[k] is None or len(arrays[k])==0]
if missing:
    raise RuntimeError(f"Required arrays missing: {missing}. Set force_load_npy=True and provide .npy files, or run model evaluation in the notebook then re-run this cell. Current arrays found: {list(k for k in arrays.keys())}")

# Optional arrays
optional = ["y_pred_sliding_2021", "y_pred_finetune_2021"]
for k in optional:
    if k not in arrays or arrays[k] is None:
        arrays[k] = None
        print(f"Optional array {k} not found - results will show NaN for these.")

# ---------------- Align lengths (if any mismatch) ----------------
# We prefer not to silently drop data; but metrics require equal lengths.
def align_pair(y_true, y_pred, name):
    if len(y_true) == len(y_pred):
        return y_true, y_pred
    # Trim to minimum length (explicit warning)
    minlen = min(len(y_true), len(y_pred))
    warnings.warn(f"Length mismatch for {name}: y_true={len(y_true)}, y_pred={len(y_pred)}. Trimming to {minlen} (min).", UserWarning)
    return y_true[:minlen], y_pred[:minlen]

# Align baseline pairs
arrays["y_true_2016"], arrays["y_pred_static_2016"] = align_pair(arrays["y_true_2016"], arrays["y_pred_static_2016"], "Static (2016)")
arrays["y_true_2021"], arrays["y_pred_static_2021"] = align_pair(arrays["y_true_2021"], arrays["y_pred_static_2021"], "Static (2021)")

# If optional present, align them to y_true_2021
for k in optional:
    if arrays.get(k) is not None:
        arrays["y_true_2021"], arrays[k] = align_pair(arrays["y_true_2021"], arrays[k], k)

# ---------------- Save arrays for reproducibility ----------------
for k, fname in npy_files.items():
    arr = arrays.get(k)
    if arr is not None:
        try:
            np.save(fname, arr)
        except Exception:
            pass

# ---------------- Compute metrics ----------------
results = {}
# static train 2016
results["Static (train 2016)"] = compute_scores(arrays["y_true_2016"], arrays["y_pred_static_2016"], labels=labels, label_names=label_names)
# static test 2021
results["Static (test 2021)"] = compute_scores(arrays["y_true_2021"], arrays["y_pred_static_2021"], labels=labels, label_names=label_names)
# optional sliding/fine-tune
if arrays.get("y_pred_sliding_2021") is not None:
    results["Sliding-window (test 2021)"] = compute_scores(arrays["y_true_2021"], arrays["y_pred_sliding_2021"], labels=labels, label_names=label_names)
else:
    results["Sliding-window (test 2021)"] = None
if arrays.get("y_pred_finetune_2021") is not None:
    results["Fine-tuned (test 2021)"] = compute_scores(arrays["y_true_2021"], arrays["y_pred_finetune_2021"], labels=labels, label_names=label_names)
else:
    results["Fine-tuned (test 2021)"] = None

# ---------------- Comparative table ----------------
rows = []
baseline_acc = results["Static (test 2021)"]["accuracy"] if results["Static (test 2021)"] is not None else np.nan

for name, res in results.items():
    if res is None:
        rows.append({
            "Model": name,
            "Accuracy": np.nan,
            "Precision_Down": np.nan, "Recall_Down": np.nan, "F1_Down": np.nan,
            "Precision_Up": np.nan, "Recall_Up": np.nan, "F1_Up": np.nan,
            "Macro_Precision": np.nan, "Macro_Recall": np.nan, "Macro_F1": np.nan,
            "Weighted_F1": np.nan,
            "Improvement_vs_static_test(%)": np.nan
        })
    else:
        prec = res["precision_per_class"]
        rec = res["recall_per_class"]
        f1 = res["f1_per_class"]
        row = {
            "Model": name,
            "Accuracy": res["accuracy"],
            "Precision_Down": prec[0], "Recall_Down": rec[0], "F1_Down": f1[0],
            "Precision_Up": prec[1], "Recall_Up": rec[1], "F1_Up": f1[1],
            "Macro_Precision": res["precision_macro"],
            "Macro_Recall": res["recall_macro"],
            "Macro_F1": res["f1_macro"],
            "Weighted_F1": res["f1_weighted"],
            "Improvement_vs_static_test(%)": ((res["accuracy"] - baseline_acc) / baseline_acc * 100) if (not np.isnan(baseline_acc) and baseline_acc != 0) else np.nan
        }
        rows.append(row)

df_compare = pd.DataFrame(rows)
cols = ["Model", "Accuracy",
        "Precision_Down", "Recall_Down", "F1_Down",
        "Precision_Up", "Recall_Up", "F1_Up",
        "Macro_Precision", "Macro_Recall", "Macro_F1",
        "Weighted_F1", "Improvement_vs_static_test(%)"]
df_compare = df_compare[cols]
pd.options.display.float_format = '{:0.4f}'.format

# ---------------- Print reports ----------------
print("\n--- Detailed classification reports ---\n")
for name, res in results.items():
    print(f"== {name} ==")
    if res is None:
        print("No predictions available for this model. Skipping classification report.\n")
    else:
        print(res["classification_report_text"])
        print()

print("\n--- Comparative table ---\n")
print(df_compare.to_string(index=False))

# ---------------- Save outputs ----------------
df_compare.to_csv("comparison_table_metrics.csv", index=False)
with open("comparison_results_summary.json", "w") as f:
    json.dump({k: (v["accuracy"] if v else None) for k, v in results.items()}, f, indent=2)

print("\nSaved 'comparison_table_metrics.csv' and 'comparison_results_summary.json' to the current working directory.")
print("Saved available .npy arrays (see npy_files mapping).")
```
